
#  Flats Price Analysis in Bangalore

### Web scraping is a method to obtain large amounts of data from websites

```python
# Importing required packages
import pandas as pd # For creating a dataframe
import numpy as np # For filling the missing values with nan
import requests # For requestig a web page to retrive the information
from bs4 import BeautifulSoup # it allows us to pull data out HTML documents
```

```python
#getting data
#urls=[]
price=[]
area=[]
price_sqft=[]
builder=[]
unit=[]
building_status=[]
location=[]
flat=[]
for i in range(2,120):
    a="https://www.makaan.com/bangalore-residential-property/buy-property-in-bangalore-city?page="+str(i)+"&_=1668148177035"
    #urls.append(a)
    page=requests.get(a)
    soup=BeautifulSoup(page.text)
    b=soup.find_all('a',attrs={'class':'seller-name'})
    for i in b:
        builder.append(i.text)
    per_sqft=soup.find_all('td',attrs={'class':'lbl rate'})
    for i in per_sqft:
        price_sqft.append(i.text)
    a=soup.find_all('td',attrs={'class':'size'})
    for i in a:
        area.append(i.text)
    u=soup.find_all('span',attrs={'class':'unit'})
    for i in u:
        unit.append(i.text)
    p=soup.find_all('span',attrs={'class':'val','itemprop':'offers'})
    for i in p:
        price.append(i.text)
    loc=soup.find_all('span',attrs={'itemprop':'addressLocality'})
    for i in loc:
        location.append((i).text)
    con_status=soup.find_all('td',attrs={'class':'val'})
    for i in con_status:
            building_status.append(i.text)
    h=[]
    flat1=soup.find_all('div',class_="title-line")
    for i in flat1:
        h.append(i.find("span",class_="val"))
    for i in h:
        if i is None:
            flat.append(np.NaN)
        else:
            flat.append(i.text)
```
